```r
# Cargamos paquetes

library(repr)
library(dplyr)

library(abind)
library(loadeR)
library(transformeR)
library(visualizeR)
library(downscaleR)
library(climate4R.UDG)
library(climate4R.climdex)
library(climate4R.indices)
library(easyVerification)

library(lattice)
library(magrittr)
library(gridExtra)
library(RColorBrewer)

# biasCorrection_RM
source("../load_bc_functions.R")
```

```r
# Region de estudio

lon = c(-10, 20)
lat = c(35,46)

# Color
color = colorRampPalette(rev(brewer.pal(n = 9, "RdYlBu")))
```

```r
# Cargo datos
tas_modelo_0 = readRDS("DatosVid/tas_botritis_model_vid_0.rds")
tas_modelo_1 = readRDS("DatosVid/tas_botritis_model_vid_1.rds")
tas_modelo_2 = readRDS("DatosVid/tas_botritis_model_vid_2.rds")
tas_modelo_3 = readRDS("DatosVid/tas_botritis_model_vid_3.rds")

tas_obs_cel = readRDS("DatosVid/tas_botritis_obs_vid.rds")
```

```r
folds_list = lapply(1981:2022, function(x) x)

bc_tas_oidio_0 = biasCorrection_RM(
    y = tas_obs_cel,
    x = tas_modelo_0,
    newdata = tas_modelo_0,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings

bc_tas_oidio_1 = biasCorrection_RM(
    y = tas_obs_cel,
    x = tas_modelo_1,
    newdata = tas_modelo_1,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings

bc_tas_oidio_2 = biasCorrection_RM(
    y = tas_obs_cel,
    x = tas_modelo_2,
    newdata = tas_modelo_2,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings

bc_tas_oidio_3 = biasCorrection_RM(
    y = tas_obs_cel,
    x = tas_modelo_3,
    newdata = tas_modelo_3,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings
```

```r
saveRDS(bc_tas_oidio_0, file = "bc_tas_botritis_vid_0.rds")
saveRDS(bc_tas_oidio_1, file = "bc_tas_botritis_vid_1.rds")
saveRDS(bc_tas_oidio_2, file = "bc_tas_botritis_vid_2.rds")
saveRDS(bc_tas_oidio_3, file = "bc_tas_botritis_vid_3.rds")
```
