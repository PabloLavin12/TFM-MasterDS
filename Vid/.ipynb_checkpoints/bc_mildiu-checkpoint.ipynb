## Configuración

```r
# Cargamos paquetes

library(repr)
library(dplyr)

library(abind)
library(loadeR)
library(transformeR)
library(visualizeR)
library(downscaleR)
library(climate4R.UDG)
library(climate4R.climdex)
library(climate4R.indices)
library(easyVerification)
library(lubridate)

library(lattice)
library(magrittr)
library(gridExtra)
library(RColorBrewer)

# biasCorrection_RM
source("../load_bc_functions.R")
```

```r
# Cargo datos: temperatura media (con sesgo y sin sesgo) y humedad relativa
tas_modelo_0 = readRDS("DatosVid/tasmin_mildiu_model_vid_0.rds")
tas_modelo_1 = readRDS("DatosVid/tasmin_mildiu_model_vid_1.rds")
tas_modelo_2 = readRDS("DatosVid/tasmin_mildiu_model_vid_2.rds")
tas_modelo_3 = readRDS("DatosVid/tasmin_mildiu_model_vid_3.rds")

tas_obs = readRDS("DatosVid/tasmin_mildiu_obs_vid.rds")
```

```r
folds_list = lapply(1981:2021, function(x) x)

bc_tas_0 = biasCorrection_RM(
    y = tas_obs,
    x = tas_modelo_0,
    newdata = tas_modelo_0,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings

bc_tas_1 = biasCorrection_RM(
    y = tas_obs,
    x = tas_modelo_1,
    newdata = tas_modelo_1,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings

bc_tas_2 = biasCorrection_RM(
    y = tas_obs,
    x = tas_modelo_2,
    newdata = tas_modelo_2,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings

bc_tas_3 = biasCorrection_RM(
    y = tas_obs,
    x = tas_modelo_3,
    newdata = tas_modelo_3,
    method = "eqm",
    extrapolation = "constant",
    cross.val = 'kfold',
    folds = folds_list) %>% suppressMessages %>% suppressWarnings
```

```r
saveRDS(bc_tas_0, file = "bc_tasmin_mildiu_vid_0.rds")
saveRDS(bc_tas_1, file = "bc_tasmin_mildiu_vid_1.rds")
saveRDS(bc_tas_2, file = "bc_tasmin_mildiu_vid_2.rds")
saveRDS(bc_tas_3, file = "bc_tasmin_mildiu_vid_3.rds")
```
